```python
from sklearn import preprocessing
LabelEncoder = preprocessing.LabelEncoder()
company_data['ShelveLoc'] = LabelEncoder.fit_transform(company_data['ShelveLoc'])
company_data['Urban'] = LabelEncoder.fit_transform(company_data['Urban'])
company_data['US'] = LabelEncoder.fit_transform(company_data['US'])
```

```python
company_data
```

```python
company_data.dtypes
```

```python
company_data['Sales'] = company_data['Sales'].astype('int')
```

```python
company_data.dtypes
```

```python
X = company_data.drop(labels='Sales', axis=1)
y = company_data[['Sales']]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(criterion='gini', max_depth=3)
dt_model.fit(X_train,y_train)
```

```python
from sklearn.tree import plot_tree
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(20,16))
plot_tree(dt_model,rounded=True,filled=True)
plt.show()
```

```python
y_train_pred = dt_model.predict(X_train)
```

```python
y_test_pred = dt_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_train,y_train_pred)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_train,y_train_pred))

```

```python
from sklearn.metrics import classification_report
print(classification_report(y_train,y_train_pred))
```

```python
accuracy_score(y_test,y_test_pred)
```

```python
print(confusion_matrix(y_test,y_test_pred))
```

```python
print(classification_report(y_test,y_test_pred))
```
